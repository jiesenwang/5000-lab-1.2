## Python Code Example

Let's look at some Python code and a plot.

```python
# Data for plotting
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)

# Creating the plot
plt.plot(x, y, label='Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cosine Wave Plot')
plt.legend()

# Display the plot
plt.show()
```
